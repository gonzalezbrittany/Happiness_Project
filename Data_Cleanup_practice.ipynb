```python
# Import library
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt # Visualization
import seaborn as sns #Visualization
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] =14
plt.rcParams['font.weight']= 'bold'
plt.style.use('seaborn-whitegrid')
```

```python
# Import dataset
# The happiness data files 2021 and trending
happiness_t_file= "Data/world-happiness-report.csv"
happiness_file = "Data/world-happiness-report-2021_test.csv"
covid_data = "Data/kaggle-covid/worldometer_data_test.csv"
screen_time = "Data/Clean_Average_Screen_Time_per_Country_2021_test.csv"
population_density = "Data/Country_Population_Land_Area_Density_test.csv"
alcohol_consumption = "Data/alcohol_consumption_bycountry_test.csv"
meat_consumption = "Data/Meat_consumption_test.csv"
median_age = "Data/Median_age_test.csv"
suicide_rate = "Data/Suicide_rate_test.csv"
```

```python
happiness_t_file_df = pd.read_csv(happiness_t_file, index_col=0)
covid_data_df = pd.read_csv(covid_data, index_col=0)
screen_time_df = pd.read_csv(screen_time, index_col=0)
population_density_df = pd.read_csv(population_density, index_col=0)
alcohol_consumption_df = pd.read_csv(alcohol_consumption, index_col=0)
meat_consumption_df = pd.read_csv(meat_consumption, index_col=0)
median_age_df = pd.read_csv(median_age, index_col=0)
suicide_rate_df = pd.read_csv(suicide_rate, index_col=0)
```

```python
happiness_covid_df = happiness_t_file_df.merge(covid_data_df , on='Country name', how = 'left')
happiness_covid_screen_df = happiness_covid_df.merge(screen_time_df , on='Country name', how = 'left')
happiness_covid_screen_density_df = happiness_covid_screen_df.merge(population_density_df , on='Country name', how = 'left')
happiness_covid_screen_density_alc_df = happiness_covid_screen_density_df.merge(alcohol_consumption_df , on='Country name', how = 'left')
happiness_covid_screen_density_alc_meat_df = happiness_covid_screen_density_alc_df.merge(meat_consumption_df , on='Country name', how = 'left')
happiness_covid_screen_density_alc_meat_age_df = happiness_covid_screen_density_alc_meat_df .merge(median_age_df , on='Country name', how = 'left')
```

```python
happiness_all_df = happiness_covid_screen_density_alc_meat_age_df .merge(suicide_rate_df , on='Country name', how = 'left')
```
